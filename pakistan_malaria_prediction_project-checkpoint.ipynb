```python
import pandas as pd
import numpy as np # Adding numpy import for better practice, though not strictly required yet

# --- UPDATED DATA LOADING ---
# FIX: Using 'r' before the string to make it a raw string, preventing the Unicode error.
try:
    # Load the first dataset (features like rain, temp, humid, population, etc.)
    df_features = pd.read_csv(r'C:\Users\Aathira baskaran\Downloads\features.csv')

    # Load the second dataset (the target variable: cases)
    df_cases = pd.read_csv(r'C:\Users\Aathira baskaran\Downloads\cases.csv')

    print("Data successfully loaded into df_features and df_cases.")

except FileNotFoundError:
    print("ERROR: One or both of the CSV files were not found.")
    print("Please check the full path and file name for both files.")
    raise

# 1. Ensure 'date' columns are in datetime format
df_features['date'] = pd.to_datetime(df_features['date'])
df_cases['date'] = pd.to_datetime(df_cases['date'])

# 2. Merge the two DataFrames on 'date' and 'district'
df_merged = pd.merge(
    df_features,
    df_cases,
    on=['date', 'district'],
    how='inner'
)

# 3. Sort by district and date (crucial for time-series feature engineering)
df_merged = df_merged.sort_values(by=['district', 'date']).reset_index(drop=True)

print("\n--- Merged Data Head ---")
print(df_merged.head())
print(f"\nTotal rows after merge: {len(df_merged)}")

# --- Step 2: Feature Engineering (Lagged Features) ---
df = df_merged.copy() 

lag_cols = ['cases', 'rain', 'temp', 'humid']
lag_periods = [7, 30]

for district in df['district'].unique():
    mask = df['district'] == district
    for col in lag_cols:
        for lag in lag_periods:
            new_col_name = f'{col}_lag_{lag}'
            # Apply shift only to the current district's data
            df.loc[mask, new_col_name] = df.loc[mask, col].shift(lag)

# Drop rows with NaN values (the first 30 days of each district's data will be dropped)
df_model = df.dropna().reset_index(drop=True)

print("\n--- Data Head with Lagged Features ---")
print(df_model.head())
print(f"\nRows after dropping NaNs: {len(df_model)}")


# --- Step 3: Defining the Target Variable (Outbreak Hotspot) ---

# Calculate the 90th percentile of cases for each district
district_percentiles = df_model.groupby('district')['cases'].quantile(0.90).reset_index()
district_percentiles.rename(columns={'cases': 'case_threshold'}, inplace=True)

# Merge the threshold back to the main DataFrame
df_model = pd.merge(df_model, district_percentiles, on='district', how='left')

# Create the binary target variable 'is_hotspot'
df_model['is_hotspot'] = (df_model['cases'] >= df_model['case_threshold']).astype(int)

# Check the balance of the target variable
hotspot_balance = df_model['is_hotspot'].value_counts(normalize=True)

print("\n--- Hotspot Target Variable Balance (0=No Outbreak, 1=Hotspot) ---")
print(hotspot_balance)
print(df_model[['district', 'date', 'cases', 'is_hotspot']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import numpy as np

# --- 1. Define Features (X) and Target (y) ---

# All columns that the model can use to predict the future.
# We exclude the original 'cases' (since it's in the future of the lagged features),
# the target 'is_hotspot', the 'date', 'district', and the 'case_threshold'.
feature_cols = [
    # Current Features
    'rain', 'temp', 'humid', 'holiday_flag', 'itn', 'irs', 'population',
    # Lagged Features (The most powerful predictors!)
    'cases_lag_7', 'cases_lag_30', 'rain_lag_7', 'rain_lag_30', 
    'temp_lag_7', 'temp_lag_30', 'humid_lag_7', 'humid_lag_30'
]

X = df_model[feature_cols]
y = df_model['is_hotspot']

# --- 2. Time-Series Split ---

# Determine the split date. A common approach is to reserve the last 15-20% of the data for testing.
# Let's use the last 20% of the entire dataset as the test set.

# Find the cutoff index for the last 20% of rows
split_index = int(len(df_model) * 0.8)

# Split the data sequentially: Train on older data, Test on newer data
X_train = X.iloc[:split_index]
X_test = X.iloc[split_index:]
y_train = y.iloc[:split_index]
y_test = y.iloc[split_index:]

print(f"Total data points: {len(X)}")
print(f"Training data points (older dates): {len(X_train)}")
print(f"Testing data points (newer dates): {len(X_test)}")
print(f"First test date: {df_model.iloc[split_index]['date']}")
print("-" * 30)

# --- 3. Model Training (Random Forest) ---

# Initialize the model. We use a Random Forest because it handles complex relationships well
# and doesn't require extensive feature scaling like linear models.
# class_weight='balanced' is CRITICAL because the target (is_hotspot) is imbalanced (10% vs 90%)
model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

print("Training Random Forest Classifier...")
model.fit(X_train, y_train)
print("Training complete.")

# --- 4. Prediction and Evaluation ---

# Predict probabilities on the test set
y_pred_proba = model.predict_proba(X_test)[:, 1] # Probability of being a hotspot (class 1)
y_pred = model.predict(X_test)

# Evaluation
print("\n--- Model Evaluation (Predicting Future Outbreaks) ---")
print(classification_report(y_test, y_pred))

# AUC-ROC Score (a robust metric for imbalanced classification)
auc_roc = roc_auc_score(y_test, y_pred_proba)
print(f"AUC-ROC Score on Test Set: {auc_roc:.4f}")

# Feature Importance (helps understand which factors drive the prediction)
feature_importances = pd.Series(model.feature_importances_, index=feature_cols).sort_values(ascending=False)
print("\n--- Top 5 Feature Importances ---")
print(feature_importances.head())
```

```python
# Create a DataFrame of the test set dates and districts
test_data_for_prediction = df_model.iloc[split_index:].copy()

# Add the predicted probability of being a hotspot
test_data_for_prediction['hotspot_proba'] = y_pred_proba

# Set a stricter probability threshold (e.g., predict as hotspot if proba > 0.4)
# Note: We are using a lower threshold than the default 0.5 to increase Recall.
prediction_threshold = 0.40 
test_data_for_prediction['predicted_hotspot'] = (test_data_for_prediction['hotspot_proba'] >= prediction_threshold).astype(int)

# --- Find the Top 5 Riskiest Districts in the latest data ---
# Filter for predicted hotspots and calculate the average risk per district
risky_districts = test_data_for_prediction[test_data_for_prediction['predicted_hotspot'] == 1]
hotspot_report = risky_districts.groupby('district')['hotspot_proba'].agg(
    Average_Risk_Score='mean',
    Predicted_Hotspot_Days='count'
).sort_values(by='Average_Risk_Score', ascending=False)

print("\n--- Actionable Prediction: Top Riskiest Districts (Predicted Hotspots) ---")
print(hotspot_report.head(5))
```

```python
import pandas as pd
from datetime import timedelta

# NOTE: The df_model variable and the trained 'model' object from Step 4 
# are required and assumed to be available in your environment.

# --- 1. Define Prediction Period and Features ---
# Find the actual last date in the df_model to start the forecast from
last_date = df_model['date'].max()
start_forecast_date = last_date + timedelta(days=1)
forecast_periods = 60 # Predict for the next 60 days (approx. 2 months)
all_districts = df_model['district'].unique()

# The features required for the model (defined in Step 4)
feature_cols = [
    'rain', 'temp', 'humid', 'holiday_flag', 'itn', 'irs', 'population',
    'cases_lag_7', 'cases_lag_30', 'rain_lag_7', 'rain_lag_30', 
    'temp_lag_7', 'temp_lag_30', 'humid_lag_7', 'humid_lag_30'
]

# --- 2. Create the Future/Forecast DataFrame ---
# Create an empty list to hold forecast rows
forecast_rows = []

for district in all_districts:
    for day in range(forecast_periods):
        current_date = start_forecast_date + timedelta(days=day)
        forecast_rows.append({'district': district, 'date': current_date})

df_forecast = pd.DataFrame(forecast_rows)
df_forecast['date'] = pd.to_datetime(df_forecast['date'])

# --- 3. Simulate Future Features (The Simplest Approach) ---

# We will fill current features (rain, temp, etc.) with the mean of the last 30 days of historical data for each district.
# This serves as a baseline/simulated forecast.
df_historical_tail = df_model[df_model['date'] > (last_date - timedelta(days=30))]

for district in all_districts:
    # Calculate average current features from the last 30 historical days
    avg_features = df_historical_tail[df_historical_tail['district'] == district][
        ['rain', 'temp', 'humid', 'holiday_flag', 'itn', 'irs', 'population']
    ].mean()
    
    # Fill the forecast data with these averages
    mask = df_forecast['district'] == district
    for col in avg_features.index:
        df_forecast.loc[mask, col] = avg_features[col]
    
    # --- 4. Create Lagged Features for the Forecast ---
    # This is the tricky part: we need to append the historical tail to the forecast data
    # to calculate the lags correctly across the boundary.
    district_data = pd.concat([
        df_historical_tail[df_historical_tail['district'] == district].copy(),
        df_forecast[df_forecast['district'] == district].copy()
    ]).sort_values('date').reset_index(drop=True)
    
    # For a real forecast, you would have to sequentially predict cases (autoregression)
    # but for this example, we'll use the simplest assumption:
    # For future 'cases_lag_7' or 'cases_lag_30', we'll use the mean of the lagged cases from the historical tail.
    
    for col_lag in ['cases_lag_7', 'cases_lag_30', 'rain_lag_7', 'rain_lag_30', 
                    'temp_lag_7', 'temp_lag_30', 'humid_lag_7', 'humid_lag_30']:
        
        # Calculate the mean of the lagged feature from the historical tail
        avg_lag = df_historical_tail[df_historical_tail['district'] == district][col_lag].mean()
        
        # Fill the forecast period with this average lagged value
        df_forecast.loc[mask, col_lag] = avg_lag


# --- 5. Prediction ---

# Ensure the DataFrame has only the features required for prediction
X_forecast = df_forecast[feature_cols]

# Predict probabilities using the trained model
df_forecast['hotspot_proba'] = model.predict_proba(X_forecast)[:, 1]

# --- 6. Generate User-Friendly Output ---

# Aggregate the risk across the entire 2-month forecast period
forecast_summary = df_forecast.groupby('district')['hotspot_proba'].agg(
    Average_Risk_Score='mean',
    Days_Above_50_Percent_Risk=lambda x: (x > 0.50).sum()
).sort_values(by='Average_Risk_Score', ascending=False)

# Identify the overall top risk factors from the Feature Importance list (from Step 4)
top_factors = ['rain_lag_30', 'humid_lag_30', 'rain_lag_7', 'temp_lag_30'] 
top_factor_string = ", ".join([f"High {col.replace('_', ' ')}" for col in top_factors])


# --- Final Output Generation ---
top_district = forecast_summary.iloc[0]
top_district_name = forecast_summary.index[0]

print("\n" * 2)
print("====================================================================================")
print(f"| MALARIA OUTBREAK FORECAST: {start_forecast_date.strftime('%B %Y')} to {(start_forecast_date + timedelta(days=forecast_periods - 1)).strftime('%B %Y')} |")
print("====================================================================================")

print(f"\nðŸ’¡ HIGHEST RISK HOTSPOT: **{top_district_name.upper()}**")
print(f"   - **Forecasted Risk Score:** {top_district['Average_Risk_Score']:.2f} (Average probability of outbreak per day)")
print(f"   - **High-Risk Days:** {int(top_district['Days_Above_50_Percent_Risk'])} out of {forecast_periods} days have > 50% risk.")

print("\n**WHY THIS DISTRICT IS AT RISK:**")
print(f"Your predictive model indicates that the primary drivers of risk are related to **climate and delayed effects**.")
print(f"The top factors influencing this forecast are: **{top_factor_string}**. ")
print("Specifically, the heavy rainfall and high humidity recorded **30 days prior to the forecast period** are strongly correlated with an increase in cases now, due to the mosquito breeding and incubation cycle.")

print("\n**TOP 5 RISKY DISTRICTS FOR IMMEDIATE INTERVENTION:**")
print(forecast_summary.head(5).to_markdown())

print("\n====================================================================================")
print("NOTE: This prediction assumes future features will follow recent historical averages.")
print("For a true real-time forecast, update the current features (rain, temp, etc.) with")
print("actual weather forecasts and new intervention data.")
print("====================================================================================")
```

```python
import pandas as pd
from datetime import timedelta

# NOTE: We assume 'df_model' (your fully engineered data) and the trained 'model' 
# object are available from the previous steps.

# --- 1. Define Prediction Period ---
last_date = df_model['date'].max()
start_forecast_date = last_date + timedelta(days=1)

# Define the 6-month period for the forecast (August 2025 to January 2026)
forecast_start_month = 8  # August
forecast_start_year = start_forecast_date.year 
forecast_periods_days = 184 # Approx. 6 months
all_districts = df_model['district'].unique()

# --- 2. Calculate Monthly Historical Averages ---

# Prepare historical data to get monthly averages
df_model['month'] = df_model['date'].dt.month

# Calculate the average current features for each month and district across all years
monthly_avg_features = df_model.groupby(['district', 'month'])[[
    'rain', 'temp', 'humid', 'holiday_flag', 'itn', 'irs', 'population'
]].mean().reset_index()

# Calculate the average lagged features for each month and district
monthly_avg_lags = df_model.groupby(['district', 'month'])[[
    'cases_lag_7', 'cases_lag_30', 'rain_lag_7', 'rain_lag_30', 
    'temp_lag_7', 'temp_lag_30', 'humid_lag_7', 'humid_lag_30'
]].mean().reset_index()


# --- 3. Create the Future/Forecast DataFrame and Populate Features ---
forecast_rows = []
current_date = start_forecast_date

for i in range(forecast_periods_days):
    for district in all_districts:
        forecast_month = current_date.month
        
        # Look up the historical average for this district and month
        avg_feats = monthly_avg_features[
            (monthly_avg_features['district'] == district) & 
            (monthly_avg_features['month'] == forecast_month)
        ]
        
        avg_lags = monthly_avg_lags[
            (monthly_avg_lags['district'] == district) & 
            (monthly_avg_lags['month'] == forecast_month)
        ]
        
        # Create the row data (using historical monthly average as the feature value)
        row_data = {
            'district': district, 
            'date': current_date, 
            **avg_feats.drop(['district', 'month'], axis=1).iloc[0].to_dict(),
            **avg_lags.drop(['district', 'month'], axis=1).iloc[0].to_dict()
        }
        forecast_rows.append(row_data)
    
    current_date += timedelta(days=1)

df_forecast_6m = pd.DataFrame(forecast_rows)
df_forecast_6m.drop(columns=['month'], errors='ignore', inplace=True)

# --- 4. Prediction ---
feature_cols = df_forecast_6m.columns.drop(['district', 'date']).tolist()
X_forecast = df_forecast_6m[feature_cols]

# Predict probabilities using the trained model
df_forecast_6m['hotspot_proba'] = model.predict_proba(X_forecast)[:, 1]


# --- 5. Generate User-Friendly Report ---

# Aggregate the risk by district and by month
df_forecast_6m['forecast_month'] = df_forecast_6m['date'].dt.strftime('%Y-%m')

# Summary of overall risk for the 6 months
overall_risk_summary = df_forecast_6m.groupby('district')['hotspot_proba'].agg(
    Average_Risk_Score='mean',
    High_Risk_Days=lambda x: (x > 0.50).sum()
).sort_values(by='Average_Risk_Score', ascending=False)

# Monthly breakdown for the top 5 riskiest districts
top_5_districts = overall_risk_summary.head(5).index
monthly_risk_breakdown = df_forecast_6m[df_forecast_6m['district'].isin(top_5_districts)].groupby(
    ['district', 'forecast_month']
)['hotspot_proba'].mean().unstack(level=1).fillna(0).applymap(lambda x: f"{x*100:.1f}%")

# Identify the overall top risk factors (from the Feature Importance list in Step 4)
top_factors = ['Rainfall (30 days ago)', 'Humidity (30 days ago)', 'Rainfall (7 days ago)', 'Temperature (30 days ago)'] 
top_factor_string = ", ".join(top_factors)

# --- Final Output ---

print("\n" * 2)
print("==========================================================================================")
print("| **MALARIA OUTBREAK HOTSPOT FORECAST REPORT: AUG 2025 - JAN 2026** (6 MONTHS) |")
print("==========================================================================================")
print(f"**Forecast Period:** {start_forecast_date.strftime('%B %Y')} to {current_date.strftime('%B %Y')}")
print(f"**Methodology:** Model trained on historical data. Future features are based on **Historical Seasonal Averages**.")

print("\n### ðŸ¥‡ Six-Month Overall Highest Risk Districts (Top 5)")
print("These districts require prioritized resource allocation throughout the forecast period.")
print(overall_risk_summary.head(5).to_markdown())

print("\n### ðŸš¨ Key Takeaway: Primary Cause of Risk")
print(f"Your model highlights that the highest risk is primarily driven by **lagged climatic factors**.")
print(f"The most important risk indicators are: **{top_factor_string}**.  This confirms that preventive measures must be taken 3-5 weeks *before* the expected peak in cases.")

print("\n### ðŸ“‰ Monthly Risk Breakdown for Top Hotspots (Average Daily Outbreak Probability)")
print("This matrix shows the peak risk months for targeted intervention:")
print(monthly_risk_breakdown.to_markdown())

print("\n--- Actionable Recommendation ---")
print(f"**FOCUS:** Districts like **{overall_risk_summary.index[0]}** show the highest average risk. Health officials should mobilize teams to these areas, coordinating activity 30 days after peak monsoon rainfall.")
print("The high risk in the **lag_30** features strongly suggests that interventions like **IRS/ITN distribution** should be planned well in advance of the forecast period.")
```

```python
# --- Corrected Step: Recalculate Monthly Risk Breakdown ---

# The column df_forecast_6m['forecast_month'] already holds 'YYYY-MM' strings, 
# which is perfect for grouping.

# Group by district and the forecast month, then count the number of days flagged as 'predicted_hotspot'
monthly_risk_breakdown_counts = df_forecast_6m.groupby(
    ['district', 'forecast_month']
)['predicted_hotspot'].sum().unstack(level=1).fillna(0)

# Get the month names for the final display
month_names = {i: pd.to_datetime(c).strftime('%b %Y') for i, c in enumerate(monthly_risk_breakdown_counts.columns)}

# Calculate the actual percentage of days flagged as a hotspot (Count / Total Days in Month)
def calculate_risk_percentage(row):
    percentages = []
    for col in monthly_risk_breakdown_counts.columns:
        # Determine the number of days in the specific month (approx 30 or 31)
        # Using a fixed 30 for simplicity, or we can look up the actual days in the forecast date range
        # For simplicity in this fixed forecast, we'll use 30 days as a safe approximation.
        days_in_month = 30 
        
        count = row[col]
        risk_percentage = (count / days_in_month) * 100
        percentages.append(f"{min(risk_percentage, 100):.0f}%") # Cap at 100% and format as integer percentage
    return percentages

# Apply the calculation
monthly_risk_breakdown_tuned = monthly_risk_breakdown_counts.apply(calculate_risk_percentage, axis=1, result_type='expand')

# Apply the clean month names
monthly_risk_breakdown_tuned.columns = [f"Risk in {month_names[i]}" for i in range(len(monthly_risk_breakdown_counts.columns))]

# Filter down to the top 5 districts
monthly_risk_breakdown_tuned = monthly_risk_breakdown_tuned.loc[top_5_districts]

print("\n### ðŸ“ˆ Monthly Hotspot Risk Breakdown (Percentage of Days Flagged as Hotspot) - CORRECTED")
print("This matrix shows the peak risk months for targeted, time-sensitive intervention:")
print(monthly_risk_breakdown_tuned.to_markdown(index=True))
```
